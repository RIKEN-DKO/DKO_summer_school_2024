````python
import json

# Function to extract the 'system', 'user', and 'assistant' content
def extract_messages(jsonl_file):
    extracted_data = []
    
    with open(jsonl_file, 'r') as file:
        for line in file:
            # Load each line as a JSON object
            entry = json.loads(line)
            messages = entry['messages']
            
            # Extract the content of 'system', 'user', and 'assistant'
            system_message = next((msg['content'] for msg in messages if msg['role'] == 'system'), None)
            user_message = next((msg['content'] for msg in messages if msg['role'] == 'user'), None)
            assistant_message = next((msg['content'] for msg in messages if msg['role'] == 'assistant'), None)
            
            extracted_data.append([system_message, user_message, assistant_message])
    
    return extracted_data

# Example usage
jsonl_file = 'test_data_2.jsonl'
import numpy as np
from nltk.translate.bleu_score import sentence_bleu
from rouge_score import rouge_scorer
from nltk.translate.meteor_score import meteor_score
import difflib
from openai import OpenAI

# Extract data from the JSONL file
data = extract_messages(jsonl_file)

# Initialize OpenAI client
client = OpenAI()

final_data=[]

# Iterate through the data
for row in data:
    system_message = row[0]
    user_message = row[1]
    assistant_message = row[2]

    # Get completion from the model
    completion = client.chat.completions.create(
        model="ft:gpt-4o-mini-2024-07-18:personal:juliob:A5juwrls",
        messages=[
            {"role": "system", "content": system_message},
            {"role": "user", "content": user_message}
        ]
    )
    hypothesis = completion.choices[0].message.content
    
    # Reference is the assistant's message
    reference = assistant_message
    
    final_data.append([reference,hypothesis])
    
print(final_data)
````
